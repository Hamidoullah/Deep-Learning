# Le finetuning d'un modèle de langage comme AraGPT2 sur un corpus spécifique
Le finetuning d'un modèle de langage comme AraGPT2 sur un corpus spécifique consiste à entraîner le modèle existant sur de nouvelles données adaptées à un domaine ou à un style spécifique. Voici les étapes principales pour réaliser un finetuning sur un corpus spécifique :

## Commençons par un raffinement avec un ensemble d'articles mixtes. Ces articles combinent des informations sur l'actualité.

```python
import os
os.environ["WANDB_MODE"] = "disabled"
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model = AutoModelForCausalLM.from_pretrained("aubmindlab/aragpt2-base", trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained("aubmindlab/aragpt2-base", trust_remote_code=True)

# Add a padding token to the tokenizer
if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})
    # Update the model's embedding layer to accommodate the new token
    model.resize_token_embeddings(len(tokenizer))

from datasets import load_dataset

# Charger le fichier texte
dataset = load_dataset("text", data_files={"train": "/kaggle/input/train-data/2015-07-21.txt", "validation": "/kaggle/input/train-data/test2.txt"})

# Tokeniser les données
def tokenize_function(examples):
    # Add padding='max_length' to pad all sequences to the maximum length
    # We need to create labels for the CausalLM model
    # By shifting the input_ids by one position to the right
    # we can use the input as the target for the next token prediction
    tokenized = tokenizer(examples["text"], padding='max_length', truncation=True, max_length=256)
    tokenized["labels"] = tokenized["input_ids"].copy()
    return tokenized

tokenized_datasets = dataset.map(tokenize_function, batched=True)
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=5e-5,
    weight_decay=0.01,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,  # Nombre d'époques
    save_strategy="epoch",
    logging_dir="./logs",
    logging_steps=10,
    save_total_limit=2,  # Nombre maximum de sauvegardes à conserver
    report_to="none",
)
from transformers import Trainer

# Créer le trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
)

# Lancer l'entraînement
trainer.train()

model.save_pretrained("./finetuned_aragpt2")
tokenizer.save_pretrained("./finetuned_aragpt2")
```
